```python
#matches
matches['city'].fillna('venue',inplace=True)
matches['player_of_match'].fillna('No Award',inplace=True)
matches['winner'].fillna('no winner',inplace=True)
matches['result_margin'].fillna('0',inplace=True)
matches['target_runs'].fillna('0',inplace=True)
matches['target_overs'].fillna('0',inplace=True)

#deliveries
deliveries['non_striker'].fillna('unknown',inplace=True)
deliveries['batsman_runs'].fillna('0',inplace=True)
deliveries['extra_runs'].fillna('0',inplace=True)
deliveries['total_runs'].fillna('0',inplace=True)
deliveries['extras_type'].fillna('unknown',inplace=True)
deliveries['is_wicket'].fillna('unknown',inplace=True)
deliveries['player_dismissed'].fillna('unknown',inplace=True)
deliveries['dismissal_kind'].fillna('unknown',inplace=True)
deliveries['fielder'].fillna('unknown',inplace=True)
```

```python
if matches['method'].isnull().mean() > 0.5:
  matches.drop(columns=['method'], inplace= True)
else:
  matches['method'].fillna('normal',inplace= True)
```
